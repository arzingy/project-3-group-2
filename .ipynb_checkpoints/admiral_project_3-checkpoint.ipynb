```python
import geopandas as gpd
points = gpd.read_file("Points.shp")
areas = gpd.read_file("Areas.shp")
```

```python
def convert_to_float(coord):
    # Split the coordinate into components (degrees, minutes, direction)
    parts = coord.split()
    degrees = float(parts[0])
    minutes = float(parts[1])
    direction = parts[2].upper()  # Ensure direction is uppercase
    
    # Calculate the float value (degrees + minutes/60)
    value = degrees + minutes / 60
    
    # Apply negative if direction is S or W
    if direction in ['S', 'W']:
        value = -value
    
    return round(value, 5)
```

```python
def extract_year(date):
    if date and len(date) >= 4:
        return date[:4]
    return None
```

```python
points = points[[
    'wreck_id',
    'name',
    'type',
    'flag',
    'obstructio',
    'latitude',
    'longitude',
    'water_dept',
    'date_sunk'
    ]]

points['latitude'] = points['latitude'].apply(convert_to_float)
points['longitude'] = points['longitude'].apply(convert_to_float)

points['date_sunk'] = points['date_sunk'].str.replace('"','')
points['year_sunk'] = points['date_sunk'].apply(extract_year)

points = points.rename(columns={
    'wreck_id': 'wreck_id',
    'name': 'wreck_name',
    'type': 'vessel_type',
    'obstructio': 'obstruction_type',
    'latitude': 'lat',
    'longitude': 'lon',
    'water_dept': 'water_depth'
})

points.to_csv('points.csv', index=False)

points.to_json('points.json', orient='records')
```
